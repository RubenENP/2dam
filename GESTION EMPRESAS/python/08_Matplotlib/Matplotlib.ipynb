### Scatter

Para el siguiente gráfico "scatter" utizaremos un cmap, los cmap disponibles están en la siguiente página:  
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
#gráfico de burbujas o grafico de bolas
def f(i,j):
    return (j**2)*(i**2)

x=[]
y=[]
z=[]
N=10
step=2
for i in np.arange(-N, N+step, step):
    for j in np.arange(-N, N+step, step):
        x.append(i)
        y.append(j)
        z.append(f(i,j))
        
plt.scatter(x, y, c=z, cmap='viridis',s=z, alpha=0.5)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)
plt.show()
```

Haz la gráfica del analisis de frecuencias realizado del archivo "../04_RW/el_quijote.txt"  
tienes que obtener la siguiente gráfica:  
<img src="f6.png" ALIGN="left">

```python
"""Esribe tu código aquí"""
abc="ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
archivo=open("el_quijote.txt", encoding="utf8")
quijote=archivo.read()
f=[]

frecuencia=0
letratotaleQuijote=0

for i in abc:
    frecuencia = 0
    letratotaleQuijote=0
    
    for j in quijote:
        if (j == "á" or j == "Á"):
            j = "a"
        if (j == "é" or j == "É"):
            j = "e"
        if (j == "í" or j == "Í"):
            j = "i"
        if (j == "ó" or j == "Ó"):
            j = "o"
        if (j == "ú" or j == "ü" or j == "Ú"):
            j = "u"
        
        if(i == j.upper()):
            frecuencia+=1
        if(j.upper() in abc):
            letratotaleQuijote+=1
    f.append(frecuencia)
    
x = np.array(['A','B','C','D','E','F','G','H','I','J','K','L','M','N','Ñ','O','P','Q','R','S','T','U','V','W','X','Y','Z'])
y = np.array(f)
    
plt.bar(x,y)    
plt.show()
```
